```python
def pd_plot_time_series_hcharts(df,  cfg:dict=None, mode='d3', save_img=False):
  from highcharts import Highchart
  from box import Box
  column_title = df.columns
  data = [[df[column_title[0]][i], eval(str(df[column_title[1]][i]))] for i in range(df.shape[0])]
  
  cc = Box(cfg)
  cc.name      = cc.get('name',    'my scatter')
  cc.figsize   = cc.get('figsize', (25, 15) )   
  cc.title     = 'my title'

  H = Highchart()
  H.set_options('chart', {'zoomType': 'x'})
  H.set_options('xAxis', {
                  'type': 'datetime',
                  'title': { 'text': column_title[0] }
              }) 

  H.set_options('yAxis',{ 'title': { 'text': column_title[1]}})
  H.set_options('title', {'text': cc.title})
  H.set_options('legend', {'enabled': True})
  H.add_data_set(data, 'area', 'time_series')
  
  H.set_options('plotOptions', {
            'area': {
                'fillColor': {
                    'linearGradient': { 'x1': 0, 'y1': 0, 'x2': 0, 'y2': 1},
                    'stops': [
                        [0, "Highcharts.getOptions().colors[0]"],
                        [1, "Highcharts.Color(Highcharts.getOptions().colors[0]).setOpacity(0).get('rgba')"]
                    ]},
                'marker': {
                    'radius': 2
                },
                'lineWidth': 1,
                'states': {
                    'hover': {
                        'lineWidth': 1
                    }
                },
                'threshold': None
            }
        })
  H.buildcontent()
  H.buildhtmlheader()
  html_code = H._htmlcontent.decode('utf-8')
  print(html_code)
  return html_code
```

```python
html_code = pd_plot_time_series_hcharts(df,  cfg={"title":"The grand"}, mode='d3', save_img=False)
```

```python
def pd_plot_histogram_hcharts(df,  cfg:dict=None, mode='d3', save_img=False):
  from highcharts import Highchart
  from box import Box
  column_title = df.columns
  data = [[df[column_title[0]][i], eval(str(df[column_title[1]][i]))] for i in range(df.shape[0])]
  
  cc = Box(cfg)
  cc.name      = cc.get('name',    'my scatter')
  cc.figsize   = cc.get('figsize', (25, 15) )   
  cc.title     = 'my title'

  H = Highchart()
  H.set_options('chart', {'type': 'column'})
  H.set_options('xAxis', {
                  'categories': [
                    'Jan',
                    'Feb',
                    'Mar',
                    'Apr',
                    'May',
                    'Jun',
                    'Jul',
                    'Aug',
                    'Sep',
                    'Oct',
                    'Nov',
                    'Dec'
                  ],
                  'crosshair': 'true',
                  'title': { 'text': column_title[0] }
              }) 

  H.set_options('yAxis',{'min': 0, 'title': { 'text': column_title[1]}})
  H.set_options('title', {'text': cc.title})
  H.set_options('legend', {'enabled': True})
  H.add_data_set(data, 'area')
  H.set_options('tooltip',{
      'headerFormat': '<span style="font-size:10px">{point.key}</span><table>',
    'pointFormat': '<tr><td style="color:{series.color};padding:0">{series.name}: </td>' +
      '<td style="padding:0"><b>{point.y:.1f} mm</b></td></tr>',
    'footerFormat': '</table>',
    'shared': True,
    'useHTML': True
  })
  H.set_options('plotOptions', {
    'column': {
      'pointPadding': 0,
      'borderWidth': 0,
      'groupPadding': 0,
      'shadow': False
    }
  })
  H.buildcontent()
  H.buildhtmlheader()
  html_code = H._htmlcontent.decode('utf-8')
  print(html_code)
  return html_code
```

```python
html_code = pd_plot_histogram_hcharts(df,  cfg={"title":"The grand"}, mode='d3', save_img=False)
```
